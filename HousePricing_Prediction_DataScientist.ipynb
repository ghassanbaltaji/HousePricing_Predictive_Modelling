```python
# Apply multiple linear regression to predict the house pricing as a function of the independent variables

# Divide the dataset into a dependent variable dataset and the independent variables dataset

data_independent_variables = df.drop(['price'], axis = 1)
data_dependent_variable = df['price']
```

```python
## Finding day of year function

def day_of_the_year(date_time_str):
   date_time_obj = datetime.strptime(date_time_str, '%Y-%m-%d')
   day_of_year = date_time_obj.timetuple().tm_yday

   return day_of_year
```

```python
def week_of_the_year(date_time_str):
   date_time_obj = datetime.strptime(date_time_str, '%Y-%m-%d')
   week_of_year = datetime.date(date_time_obj).isocalendar()[1]

   return week_of_year
```

```python
def month_of_the_year(date_time_str):
   date_time_obj = datetime.strptime(date_time_str, '%Y-%m-%d')
   month_of_year = date_time_obj.month

   return month_of_year

```

```python
# In our model, we consider the date to be an independent variable
# because we see a yearly periodic trend as a function of the price
# Implementing the date as a variable, it must must be numeric and more specifically a factor, so we need to convert the date into a day of the year to proceed

DAYS = []

for i in data_independent_variables['date']:
    DAYS.append(day_of_the_year(i))

data_independent_variables['DAYS'] = DAYS
data_independent_variables = data_independent_variables[['date', 'DAYS', 'spread', 'inflation', 'investments','gdp', 'production']]
data_independent_variables

```

```python
WEEK = []

for i in data_independent_variables['date']:
    WEEK.append(week_of_the_year(i))

data_independent_variables['WEEK_number'] = WEEK
data_independent_variables = data_independent_variables[['date', 'DAYS', 'WEEK_number' , 'spread', 'inflation', 'investments','gdp', 'production']]
data_independent_variables
```

```python
MONTH = []

for i in data_independent_variables['date']:
    MONTH.append(month_of_the_year(i))

data_independent_variables['MONTH_number'] = MONTH
data_independent_variables = data_independent_variables[['date', 'DAYS', 'WEEK_number' ,'MONTH_number', 'spread', 'inflation', 'investments','gdp', 'production']]
data_independent_variables
```

```python
# Change the data type of the DAYS variable

data_independent_variables['DAYS'] = data_independent_variables['DAYS'].astype('category')
data_independent_variables.dtypes
```

```python
days_df = pd.get_dummies(data_independent_variables['DAYS'], prefix = 'day', drop_first = False)
days_df
```

```python
data_independent_variables = data_independent_variables[[ 'spread', 'inflation', 'investments','gdp', 'production']]
data_independent_variables
```

```python
data_independent_variables = pd.concat([data_independent_variables, days_df ], axis = 1)
data_independent_variables
```

```python
# creating train and test sets
X_train, X_test, y_train, y_test = train_test_split(data_independent_variables,
                                                    data_dependent_variable,
                                                    test_size = 0.2,
                                                    random_state=101)
```

```python
# creating a regression model

model = LinearRegression()
```

```python
# fitting the model

model.fit(X_train,y_train)
```

```python
# making predictions

predictions = model.predict(X_test)
```

```python
# model evaluation

print('mean_squared_error : ', mean_squared_error(y_test, predictions))
print('mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

```python
# display regression coefficients and R-squared value of model

print(model.intercept_, model.coef_, model.score(data_independent_variables, data_dependent_variable))
```
